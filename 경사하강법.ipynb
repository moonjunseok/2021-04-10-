업데이트 할 W : Learning Rate * (( y예측- y실체) * X ) 평균

업데이트 할 B : Learning Rate * (( y예측 - y실체) * 1 ) 평균


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.random.rand(100)
y=0.2 * x * 0.5   # y= ax + b
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.show()   #선형 함수가 만들어지는 것을 알 수 있다.
```

```python
def plot_prediction(pred, y):
  plt.figure(figsize=(8,6))
  plt.scatter(x,y)      #실제값
  plt.scatter(x, pred)  #예측값
  plt.show()
```

## Gradient Descent 구현##

업데이트 할 W : Learning Rate * (( y예측- y실체) * X ) 평균

업데이트 할 B : Learning Rate * (( y예측 - y실체) * 1 ) 평균

```python
w = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate = 0.7 #그냥 임의로 준거임

for epoch in range(200):  # 랜덤으로 100번 정도 돌려본다
  y_pred = w * x + b

  error = np.abs(y_pred - y).mean()  # abs는 절대값을 씌워주는거
  if error < 0.001:
    break

  # gradient descent 계산,    gradient는 우리가 움직여야할 방향을 정한거다.
  w_grad = learning_rate * ((y_pred-y)*x).mean()
  b_grad = learning_rate * (y_pred-y).mean() 

  # w , b 값 갱신
  # 원래 있는 w - 움직여야할 w
  w = w - w_grad
  b = b - b_grad

  if epoch % 10 == 0:
    y_pred = w*x+b
    plot_prediction(y_pred,y)
```
